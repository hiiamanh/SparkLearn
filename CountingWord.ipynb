# Counting Word Occurrences using Flatmap()

```python
from pyspark import SparkConf, SparkContext
import re

conf = SparkConf().setMaster("local").setAppName("CountingWord")
sc = SparkContext(conf=conf)
```

```python
def normalizeWords(text):
    return re.compile(r'\W+',re.UNICODE).split(text.lower())
```

```python
book = sc.textFile("D:/LearningSpark/dataset/Book")
words = book.flatMap(normalizeWords)
wordCounts = words.map(lambda x: (x,1)).reduceByKey(lambda x,y: x + y)
wordCountsSorted = wordCounts.map(lambda xy : (xy[1], xy[0])).sortByKey(False)
wordCountsSorted.collect()
```
